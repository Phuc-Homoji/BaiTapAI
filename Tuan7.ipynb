```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dt=pd.read_csv('Salary_Data.csv')
dt.head(5)
dt.shape
sns.pairplot(dt) 
```

```python
plt.plot(dt)
```

```python
x=dt.drop(['Salary'],axis=1)
y=dt.drop(['YearsExperience'],axis=1)
plt.plot(x,y,'green',label ='Data')
plt.xlabel('Exp')
plt.ylabel('Salary')
plt.legend() 
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
Hoiquy = LinearRegression()
Hoiquy.fit(x,y)
print('Giatri Intercept: ',Hoiquy.intercept_)
print('Giatri Coefficent:',Hoiquy.coef_)
print('Phuong trinh hoi quy: y =',Hoiquy.intercept_,' + x.',Hoiquy.coef_)
```

```python
y_pred=Hoiquy.predict(x).round(1)
cal=pd.DataFrame(np.c_[y,y_pred],columns=['Original Salary','Predicted Salary'])
```

```python
cal['saiso']=abs(cal['Original Salary']-cal['Predicted Salary'])
cal.head()
```

```python
#ve do thi voi duong hoi quy
plt.scatter(x,y,color='r')
plt.plot(x,Hoiquy.predict(x),color='b')
plt.plot(x,y_pred,color='g')
plt.title('Salary vs Exp')
plt.xlabel('Year of ex')
plt.ylabel('Salary')
plt.show()
```

```python

import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# doc du lieu
data=pd.read_csv('1.01. Simple linear regression.csv')
data.head(5)
data.shape
sns.pairplot(data)

# ve do thi
plt.plot(data)

x = data.drop(['GPA'], axis=1)
y = data.drop(['SAT'], axis=1)
plt.plot(x, y, 'green', label ='Data')
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.legend() 
plt.show()

from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x, y)
print('Gia tri intercept la: ', reg.intercept_)
print('Gia tri coeffient la: ', reg.coef_)
print('Phuong trinh hoi quy la:  y = ', reg.coef_, 'x + ', reg.intercept_)

y_pred = reg.predict(x).round(1)
cal = pd.DataFrame(np.c_[x, y, y_pred, y - y_pred], columns = ['SAT', 'Original GPA', 'Predicted GPA', 'Error'])

cal

# ve do thi voi duong hoi quy
plt.scatter(x, y, color = 'red')
plt.plot(x, reg.predict(x), color = 'blue')
plt.title('GPA vs SAT')
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```

```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

data = pd.read_csv('mtcars.csv')

# to m hnh
model = LinearRegression()


Xp = pd.DataFrame([data['wt'], data['wt']**2]).T
y = data[["mpg"]]

model.fit(Xp, y)

# v  th
range = np.arange(1.5, 5.6, 0.1)
poly_range = pd.DataFrame([range, range**2]).T
y_pred = model.predict(poly_range)
data.plot(kind='scatter', x = 'wt', y='mpg',figsize=(9,9), xlim = (0,7))
plt.plot(range, y_pred, color = 'blue')
```
